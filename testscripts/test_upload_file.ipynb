```python
import asyncio
from core.dellattachments import DellAttachments

async def test_upload_sow():
    metadata = {
        "Subsystem": "AA",
        "FileVisibleExternal": "False"
    }
    
    da = DellAttachments(debug=True)
    file_id = await da.upload("test_sow_file.json", metadata=metadata)
    
    response = {   
        "Subsystem": "AA",   
        "FileVisibleExternal": "False",
        "daFileId": file_id
    }
    
    print(f"Upload successful. File ID: {file_id}")
    return response

await test_upload_sow()

# test_upload_sow()
```
